This notebook creates further plots, which are used for the thesis, but are not directly related to the Klausmeier model.

```julia
x = -2:0.01:2
p_vals = [1, 0, -1]
colors = ["#9ABCE4","#3070B3", "#0A2D57"]

plot(title="", xlabel=L"$y$", ylabel=L"$y'$",
     legend=:topright, legendfontsize=14, size=(600,400),
     xlim=(-2,2), ylim=(-3,2), xtickfontsize=10, ytickfontsize=10)

# Draw coordinate axes
hline!([0], color=:black, lw=2.0, label="")  # x-axis
vline!([0], color=:black, lw=2.0, label="")  # y-axis

i=1
# Plot the parabolas and mark intersections with y = 0
for p in p_vals
    y = p .- x.^2
    
    # adaptive label
    lbl = p > 0 ? L"$p > 0$" : p == 0 ? L"$p = 0$" : L"$p < 0$"

    plot!(x, y, label=lbl, lw=3, color=colors[i], legendfontsize = 14)
    i = i+1

    if p > 0
        x_int = [-sqrt(p), sqrt(p)]
        scatter!(x_int, [0, 0], color="#9ABCE4", marker=:circle, markerstrokewidth=0, ms=8, label="")
    elseif p == 0
        scatter!([0], [0], color="#3070B3", marker=:circle, markerstrokewidth=0, ms=8, label="")
    end
end

display(current())
```

```julia
# Parameter range
p_vals = -1:0.01:1

# Define branches
x_plus  = [sqrt(p) for p in p_vals if p >= 0]
x_minus = [-sqrt(p) for p in p_vals if p >= 0]
p_stable = [p for p in p_vals if p >= 0]  # same length as x_

# Plot
plot(p_stable, x_plus,  color=:black, lw=3, label=L"$\mathbf{y_+}$", xtickfontsize=10, ytickfontsize=10)
plot!(p_stable, x_minus, color=:black, lw=3, linestyle=:dash, label=L"$\mathbf{y_-}$", legendfontsize = 14)


# Axes and labels
vline!([0], color=:gray, lw=1, linestyle=:dot, label="")
hline!([0], color=:gray, lw=1, linestyle=:dot, label="")

scatter!([0], [0], color=:black, marker=:circle, markerstrokewidth=0, ms=8, label="")
xlabel!(L"$p$")
ylabel!("")
title!("")
xlims!(-1, 1)
ylims!(-1.5, 1.5)
xticks!(-1:1:1)
yticks!(-1:1:1)
```
