Notes on usage:
- t_fixed = true:
    - consider fixed observation time window
    - specify end time by t_end
    - M equidistant time points (i.e. M samples) are picked in given observation time window
- t_fixed = false:
    - observation time window varies with sample size M
    - specify t_step, time steps at which observations are picked
    - observation length depends on t_step and M, i.e. observation length = t_step * M
